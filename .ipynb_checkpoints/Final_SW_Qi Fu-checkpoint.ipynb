Write a GUI program for a  four function calculator. At the minimum GUI contains- 3 labels, 3 text boxes for 2 inputs and one output, 4 radio buttons for add, substract, multiply and divide and two command buttons for ‘Calculate’ and ‘Reset’. The calculator inputs are integers only. Upload source code and screen shots. 

```python
from tkinter import *
import math
from PIL import Image, ImageTk

def resizeImage(path):
    im1 = Image.open(path)
    # adjust width and height to your needs
    width = 50
    height = 50
    im1 = im1.resize((width, height), Image.ANTIALIAS)    # best down-sizing filter
    im1.save(path)
    return PhotoImage(file=path)

class calc:

    def equals(self):
        """when the equal button is pressed"""
        self.expression = self.e1.get() + self.operation + self.e2.get()
        ## Print for debuging
#         print (self.expression)
        try: 
            self.value= eval(self.expression) #evaluate the expression using the eval function
        except SyntaxError or NameErrror:
            self.e3.delete(0,END)
            self.e3.insert(0,'Invalid Input!')
        else:
            self.e3.delete(0,END)
            self.e3.insert(0,self.value)

    def clearall(self): 
        """when clear button is pressed,clears the text input area"""
        self.e1.delete(0,END)
        self.e2.delete(0,END)
        self.e3.delete(0,END)
        self.operation = ''

    def action(self,argi): 
        """pressed button's value is inserted into the end of the text area"""
        self.operation = argi
        self.e2.focus_set() #Sets focus on the input2 text area
        

    def __init__(self,master, add_img, sub_img, mul_img, div_img):
        """Constructor method"""
        master.title('Calulator_Qi Fu') 
        master.geometry()
        
        ## Create label and text block for input 1
        
        label1 = Label(master, text="Input 1:").grid(row=0, column=0)
        self.e1 = Entry(master, width = 3)
        self.e1.grid(row=0,column=3, columnspan=2,pady=2)
        self.e1.focus_set() #Sets focus on the input text area

        ## Create label and text block for input 2
        label2 = Label(master, text="Input 2:").grid(row=1, column=0)
        self.e2 = Entry(master, width = 3)
        self.e2.grid(row=1,column=3,columnspan=2,pady=2)
        
        ## Create label and text block for result
        
        label3 = Label(master, text="Result:").grid(row=2, column=0)
        self.e3 = Entry(master, width = 3)
        self.e3.grid(row=2,column=3,columnspan=2,pady=2)

        #Generating Buttons
        
        Button(master,text='Calculate',width=8,command=lambda:self.equals()).grid(row= 2, column=5)
        Button(master,text='Reset',width=8,command=lambda:self.clearall()).grid(row= 0, column=5)
        

        Radiobutton(master, image = add_img, padx = 5,command=lambda:self.action('+'), value = 1).grid(row=4, column=0)
        Radiobutton(master, image = sub_img, padx = 5,command=lambda:self.action('-'), value = 2).grid(row=4, column=1)
        Radiobutton(master, image = mul_img, padx = 5,command=lambda:self.action('*'), value = 3).grid(row=4, column=2)
        Radiobutton(master, image = div_img, padx = 5,command=lambda:self.action('/'), value = 4).grid(row=4, column=3)
        
#         Button(master,image = add_img,command=lambda:self.action('+')).grid(row=4, column=0)
#         Button(master,image = sub_img,command=lambda:self.action('-')).grid(row=4, column=1)
#         Button(master,image = mul_img,command=lambda:self.action('*')).grid(row=4, column=2)
#         Button(master,image = div_img,command=lambda:self.action('/')).grid(row=4, column=3) 
      
if __name__ == "__main__":
    root = Tk()
    add_img = resizeImage('icons/add.png')
    sub_img = resizeImage('icons/sub.png')
    mul_img = resizeImage('icons/multiple.png')
    div_img = resizeImage('icons/divide.png')
    obj=calc(root, add_img, sub_img, mul_img, div_img) #object instantiated
    root.mainloop()
```
